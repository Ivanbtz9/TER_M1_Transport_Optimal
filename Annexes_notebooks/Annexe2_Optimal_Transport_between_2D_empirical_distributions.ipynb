# Premiers pas avec le module OT

```python
# Author: Remi Flamary <remi.flamary@unice.fr>
#         Kilian Fatras <kilian.fatras@irisa.fr>
#
# License: MIT License

# sphinx_gallery_thumbnail_number = 4

import numpy as np
import numpy.linalg as npl
import matplotlib.pylab as pl
import ot
import ot.plot
```

```python
n = 50  # nb samples

mu_s = np.array([0, 0])
cov_s = np.array([[1, 0], [0, 1]])

mu_t = np.array([4, 4])
cov_t = np.array([[1, -.8], [-.8, 1]])

xs = ot.datasets.make_2D_samples_gauss(n, mu_s, cov_s)
xt = ot.datasets.make_2D_samples_gauss(n, mu_t, cov_t)

a, b = np.ones((n,)) / n, np.ones((n,)) / n  # uniform distribution on samples

# loss matrix
M = ot.dist(xs, xt)
```

```python
pl.figure(0)
pl.plot(xs[:, 0], xs[:, 1], '+b', label='Source samples')
pl.plot(xt[:, 0], xt[:, 1], 'xr', label='Target samples')
pl.legend(loc=0)
pl.title('Source and target distributions')
```

```python
G0 = ot.emd(a, b, M)


pl.figure(10)
ot.plot.plot2D_samples_mat(xs, xt, G0, c=[.5, .5, .5])
pl.plot(xs[:, 0], xs[:, 1], '+b', label='Source samples')
pl.plot(xt[:, 0], xt[:, 1], 'xr', label='Target samples')
pl.legend(loc=0)
pl.title('OT matrix with samples')
```

```python
n = 20  # nb samples

mu_s = np.array([0, 0])
cov_s = np.array([[1, 0], [0, 2]])

mu_t = np.array([3, 3])
cov_t = np.array([[1, 1], [1, 2]])

xs = ot.datasets.make_2D_samples_gauss(n, mu_s, cov_s)
xt = ot.datasets.make_2D_samples_gauss(n, mu_t, cov_t)

a, b = np.ones((n,)) / n, np.ones((n,)) / n  # uniform distribution on samples

# loss matrix
M = ot.dist(xs, xt)

pl.figure(0)
pl.plot(xs[:, 0], xs[:, 1], '+b', label='Source samples')
pl.plot(xt[:, 0], xt[:, 1], 'xr', label='Target samples')
pl.legend(loc=0)
pl.title('Source and target distributions')

G0 = ot.emd(a, b, M)


pl.figure(10)
ot.plot.plot2D_samples_mat(xs, xt, G0, c=[.5, .5, .5])
pl.plot(xs[:, 0], xs[:, 1], '+b', label='Source samples')
pl.plot(xt[:, 0], xt[:, 1], 'xr', label='Target samples')
pl.legend(loc=0)
pl.title('OT matrix with samples')
```

# Un exemple avec moins de points 

```python
n = 5  # nb samples

mu_s = np.array([4, 4]) #vecteur espérance source
cov_s = np.array([[1, 0], [0, 1]]) #matrice de covariance source 

mu_t = np.array([4, 4]) #vecteur espérance target
cov_t = np.array([[1, -.8], [-.8, 1]]) #matrice de covariance target 

xs = ot.datasets.make_2D_samples_gauss(n, mu_s, cov_s) #échantillon de n samples source
xt = ot.datasets.make_2D_samples_gauss(n, mu_t, cov_t) #échantillon de n samples target

a, b = np.ones((n,)) / n, np.ones((n,)) / n  # uniform distribution on samples

# loss matrix
M = ot.dist(xs, xt)
```

```python
print(f'xs=\n{xs}\n')
print(f'xt=\n{xt}')
```

```python
npl.norm(list(xs[0]-xt[3]),2)**2 #source en ligne et target en colonne
```

```python
M # matrice des coûts pour la norme euclidienne au carrée
```

```python
pl.figure(1)
pl.imshow(M, interpolation='nearest')
pl.title('Cost matrix M') 
#Couleurs foncées impliquent petit coût,
#couleurs claires impliquent fort coût
```

```python
a,b #mesure source et cible qui sont des uniformes ici 
```

```python
pl.figure(2)
pl.plot(xs[:, 0], xs[:, 1], '+b', label='Source samples')
pl.plot(xt[:, 0], xt[:, 1], 'xr', label='Target samples')
pl.legend(loc=0)
pl.title('Source and target distributions')
```

```python
G0 = ot.emd(a, b, M) 
#Solves the Earth Movers distance problem and returns the OT matrix

G0 #The OT_matrix qui correspond à la matrice de couplage 
```

# Des mesures source et cible qui ne sont plus des uniformes.  

```python
import numpy.random as npr

n = 3  # nb samples
m= n

mu_s = np.array([4, 4]) #vecteur espérance source
cov_s = np.array([[2, 0], [0, 2]]) #matrice de covariance source 

mu_t = np.array([4, 4]) #vecteur espérance target
cov_t = np.array([[1, -.8], [-.8, 1]]) #matrice de covariance target 

xs = ot.datasets.make_2D_samples_gauss(n, mu_s, cov_s) #échantillon de n samples source
xt = ot.datasets.make_2D_samples_gauss(m, mu_t, cov_t) #échantillon de m samples target

## Les points sources et cibles sont placés de manière aléatoire dans le plan suivant des lois gaussiennes 
## Cela ne fournit pas une mesure souce et une mesure cible dans le plan 

# loss matrix
M = ot.dist(xs, xt)

pl.figure(7)
pl.imshow(M, interpolation='nearest')
pl.title('Cost matrix M') #Couleurs foncées impliquent petit coût, couleurs claires impliquent fort coût
```

```python
a ,b = npr.random(n), npr.random(n) 
a,b = a / a.sum(), b / b.sum()
a,b
```

```python
pl.figure(5)
pl.plot(xs[:, 0], xs[:, 1], '+b', label='Source samples')
pl.plot(xt[:, 0], xt[:, 1], 'xr', label='Target samples')
pl.legend(loc=0)
pl.title('Source and target distributions')
```

```python
G0 = ot.emd(a, b, M) # r"""Solves the Earth Movers distance problem and returns the OT matrix

G0 #The OT_matrix qui correspond à la matrice de couplage
```

```python
pl.figure(6)
ot.plot.plot2D_samples_mat(xs, xt, G0, c=[.5, .5, 1])
pl.plot(xs[:, 0], xs[:, 1], '+b', label='Source samples')
pl.plot(xt[:, 0], xt[:, 1], 'xr', label='Target samples')
pl.legend(loc=0)
pl.title('OT matrix with samples')
```
